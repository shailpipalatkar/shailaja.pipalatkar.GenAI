```python
dataset = dataset[:-1]
dataset = dataset.iloc[:, :-1]
dataset
```

**STEP 1**
1. Read Dataset
2. Create parser to make dictionary for symptoms value


```python
# Define the columns that represent symptoms
symptom_columns = ['Fever', 'Cough', 'Cold']
```

```python
# Initializing an empty dictionary to store the unique symptom values
unique_symptoms = {}
```

```python
for column in symptom_columns:
    if column in dataset.columns:  # Check if the column exists before accessing
        unique_values = dataset[column].unique()
        unique_symptoms[column] = sorted(unique_values.tolist())
```
